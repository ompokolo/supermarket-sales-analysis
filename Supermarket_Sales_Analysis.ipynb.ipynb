```python
plt.figure(figsize=(8,5))

plt.hist(df["Quantity"])

plt.title("Distribution of supermarket sales")
plt.xlabel("Quantity")
plt.ylabel("Unit price")

plt.grid(True)

plt.show()
```

```python
plt.bar(df["Unit price"], df["Quantity"])
plt.title("Unit price comparison")
plt.xlabel("Unit price")
plt.ylabel("Quantity")
plt.show()
```

```python
plt.scatter(df["Tax 5%"], df["Total"])

plt.title("Tax 5% vs Total")
plt.xlabel("Tax 5%")
plt.ylabel("Total")

plt.show()
```
